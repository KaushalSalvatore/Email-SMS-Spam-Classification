```python
emails=[folder + file for file in files]
emails
```

```python
words =[]
for email in emails:
    f = open(email,encoding='latin-1')
    blob=f.read()
    words+=blob.split(" ")
```

```python
len(words)
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=""
```

```python
word_dict = Counter(words)
word_dict
```

```python
word_dict = word_dict.most_common(3000)
```

```python
# split data training and testing 
x = df.iloc[:, 1:3001]
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score,confusion_matrix,precision_score
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
gnb= GaussianNB()
mnb = MultinomialNB(force_alpha=True)
bnb = BernoulliNB()
svc = SVC(C=1.0,kernel='rbf',gamma='auto')         
rfc = RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
gnb.fit(train_x,train_y)
mnb.fit(train_x,train_y)
bnb.fit(train_x,train_y)
svc.fit(train_x,train_y)
rfc.fit(train_x,train_y)

gnb_y_pred = gnb.predict(test_x)
gnb_f1 = f1_score(test_y, gnb_y_pred)
gnb_recall = recall_score(test_y, gnb_y_pred)
gnb_roc_auc = roc_auc_score(test_y, gnb_y_pred)
gnb_as = accuracy_score(test_y, gnb_y_pred)
gnb_cm = confusion_matrix(test_y, gnb_y_pred)
gnb_ps = precision_score(test_y, gnb_y_pred)
print("GaussianNB")
print(f"F1-score of GaussianNB model: {gnb_f1:.2f}")
print(f"Recall of GaussianNB model: {gnb_recall:.2f}")
print(f"ROC AUC of GaussianNB model: {gnb_roc_auc:.2f}")
print(f"Accuracy of GaussianNB model: {gnb_as * 100:.2f}%")
print(f"confusion matrix of GaussianNB model: {gnb_cm}")
print(f"precision score of GaussianNB model: {gnb_ps * 100:.2f}%")
```

```python
bnb_y_pred = bnb.predict(test_x)
bnb_f1 = f1_score(test_y, bnb_y_pred)
bnb_recall = recall_score(test_y, bnb_y_pred)
bnb_roc_auc = roc_auc_score(test_y, bnb_y_pred)
bnb_as = accuracy_score(test_y, bnb_y_pred)
bnb_cm = confusion_matrix(test_y, bnb_y_pred)
bnb_ps = precision_score(test_y, bnb_y_pred)
print("BernoulliNB")
print(f"F1-score of BernoulliNB model: {bnb_f1:.2f}")
print(f"Recall of BernoulliNB model: {bnb_recall:.2f}")
print(f"ROC AUC of BernoulliNB model: {bnb_roc_auc:.2f}")
print(f"Accuracy of BernoulliNB model: {bnb_as * 100:.2f}%")
print(f"confusion matrix of BernoulliNB model: {bnb_cm}")
print(f"precision score of BernoulliNB model: {bnb_ps * 100:.2f}%")
```

```python
mnb_y_pred = mnb.predict(test_x)
mnb_f1 = f1_score(test_y, mnb_y_pred)
mnb_recall = recall_score(test_y, mnb_y_pred)
mnb_roc_auc = roc_auc_score(test_y, mnb_y_pred)
mnb_as = accuracy_score(test_y, mnb_y_pred)
mnb_cm = confusion_matrix(test_y, mnb_y_pred)
mnb_ps = precision_score(test_y, mnb_y_pred)
print("MultinomialNB")
print(f"F1-score of MultinomialNB model: {mnb_f1:.2f}")
print(f"Recall of MultinomialNB model: {mnb_recall:.2f}")
print(f"ROC AUC of MultinomialNB model: {mnb_roc_auc:.2f}")
print(f"Accuracy of MultinomialNB model: {mnb_as * 100:.2f}%")
print(f"confusion matrix of MultinomialNB model: {mnb_cm}")
print(f"precision score of MultinomialNB model: {mnb_ps * 100:.2f}%")
```

```python
lr = LogisticRegression(solver='lbfgs', max_iter=2000)
lr.fit(train_x, train_y)
```

```python
lr_y_pred = lr.predict(test_x)
lr_f1 = f1_score(test_y, lr_y_pred)
lr_recall = recall_score(test_y, lr_y_pred)
lr_roc_auc = roc_auc_score(test_y, lr_y_pred)
lr_as = accuracy_score(test_y, lr_y_pred)
lr_cm = confusion_matrix(test_y, lr_y_pred)
lr_ps = precision_score(test_y, lr_y_pred)
print("Logistic Regression")
print(f"F1-score of Logistic Regression model: {lr_f1:.2f}")
print(f"Recall of Logistic Regression model: {lr_recall:.2f}")
print(f"ROC AUC of Logistic Regression model: {lr_roc_auc:.2f}")
print(f"Accuracy of Logistic Regression model: {lr_as * 100:.2f}%")
print(f"confusion matrix of Logistic Regression model: {lr_cm}")
print(f"precision score of Logistic Regression model: {lr_ps * 100:.2f}%")
```
